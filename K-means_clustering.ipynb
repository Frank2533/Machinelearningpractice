```python
# Perform the necessary imports
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

USarrest = pd.read_csv("/content/drive/MyDrive/PG-DBDA _ Student Folder (1)/03 Study Materials & Assignments/12 Practical Machine Learning /Datasets/USArrests.csv",index_col=0)

scaler = StandardScaler()         # Scaling model
USarrestscaled=scaler.fit_transform(USarrest)   # fit and transform of the dataset acc to scaler

USarrestscaled = pd.DataFrame(USarrestscaled,
                          columns=USarrest.columns,
                          index=USarrest.index)       # columns hedaers and index column is set to scaled data


############################################
# We need to scale the data to get the centroids according to the distribution of the points. If we do not scale the data these distribution wont be biased and hence the clustering would be unbiased.

USarrestscaled
```

```python
clustNos = [2,3,4,5,6,7,8,9,10]     # let us assume 10 nos of clusters.
Inertia = []                        #

for i in clustNos :                 # Inertia  for each nos of clusters is calculated and stored in the list named Inertia
    model = KMeans(n_clusters=i,random_state=2022)
    model.fit(USarrestscaled)
    Inertia.append(model.inertia_)

import matplotlib.pyplot as plt     # Plot the graph between inertia and no of clusters, to get the most biased no. of clusters to be considered acc to elbow decision making.

plt.plot(clustNos, Inertia, '-o')
plt.title("Scree Plot")
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(clustNos)
plt.show()

# in the graph below we can see that elbow point is found to be 4, so it will be appropriate to consider 4 for our further calculations.
```

```python
# Create a KMeans instance with clusters: Best k model
model = KMeans(n_clusters=4,random_state=2021)

# Fit model to points
model.fit(USarrestscaled)

#model.n_init
# Determine the cluster labels of new_points: labels
labels = model.predict(USarrestscaled)


clusterID = pd.DataFrame({'ClustID':labels},index=USarrest.index)
clusteredData = pd.concat([USarrest,clusterID],
                          axis='columns')


clusteredData.sort_values('ClustID')
```

```python
clusteredData.groupby('ClustID').mean()   # Mean values of the clusters are found and variations can be seen clearly.
                                         
```
